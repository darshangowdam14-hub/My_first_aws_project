```python
#count word frequency

text = "data engg data pipeline engg"
#concerting text in small and splitting it according to the spaces
word = text.lower().split()
b = {}

for i in word:
    b[i] = b.get(i,0)+1

print(word)
print(b)
```

```python
#Remove duplicates while preserving order

num = [1,2,2,3,4,4,5,2]
seen = set()
rev = []

for i in num:
    if i not in seen:
        seen.add(i)
        rev.append(i)

print(rev)        
```

```python
#Find duplicate IDs

num = [2,2,3,5,5,6,8,9]
seen = set()
res = []

for i in num:
    if i in seen not in res:
        res.append(i)
    seen.add(i)

print(res)        
```

```python
#Find missing number

num = [1,2,3,4,6]
n=0
for i in num:
    if i>n:
        n = i
        
total_sum = n * (n+1)//2
actual = sum(num)
missing = total_sum-actual
print(missing)
```
