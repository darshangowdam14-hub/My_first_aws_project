```python
#Using Filter function in List

names = ["darshan","manu","mahadev","jhon"]

name_with_a = list(filter(lambda x: "a" in x, names))
print(name_with_a)
```

```python
numbers = [1,2,3,4,5,6]

even_num = list(filter(lambda num : num%2==0, numbers))
print(even_num)
```

```python
names = ["manu","eve","bob","darshan","car"]

names_with = list(filter(lambda name : len(name) > 3, names))
print(names_with)
```
