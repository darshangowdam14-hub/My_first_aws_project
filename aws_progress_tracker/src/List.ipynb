```python
#Using Filter function in List

names = ["darshan","manu","mahadev","jhon"]

name_with_a = list(filter(lambda x: "a" in x, names))
print(name_with_a)
```

```python
numbers = [1,2,3,4,5,6]

even_num = list(filter(lambda num : num%2==0, numbers))
print(even_num)
```

```python
names = ["manu","eve","bob","darshan","car"]

names_with = list(filter(lambda name : len(name) > 3, names))
print(names_with)
```

```python
#Multiply everthin with 10

num = [1,2,3,4,5]

ans = list(map(lambda x : x*10, num))

print(ans)
```

```python
employees = [
    {"name": "John", "salary": 50000},
    {"name": "Alice", "salary": 70000},
    {"name": "Bob", "salary": 60000}
]

result = sorted(employees, key=lambda x: x["name"])

print(result)
```

```python
# Enumarate function in list ******************

letter = ["a","b","c","d"]
new_list = []
for i in letter:
    new_list.append(i.upper())
    print(new_list)
```

```python
letter = ["a","b","c","d"]
for i, val in enumerate(letter):
    print(i, val)
```

```python
#Zip function
letter = ["a","b","c","d"]
numbers = [1,2,5,4]
for i,n in zip(letter, numbers):
    print(i, n)
    
```
